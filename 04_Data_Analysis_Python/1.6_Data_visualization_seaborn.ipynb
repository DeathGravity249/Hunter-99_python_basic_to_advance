```python
import seaborn as sns
```

```python
# Basci plotting with seaborn
tips=sns.load_dataset("tips")
tips
```

```python
# create a scatter plot
import matplotlib.pyplot as plt
sns.scatterplot(x="total_bill",y='tip',data=tips)
plt.title("Scatter data set ")
plt.grid(True)
```

```python
sns.scatterplot(
    x="total_bill",
    y="tip",
    data=tips,
    hue="sex",       # color by gender
    style="smoker",  # marker style by smoker status
    size="size",     # point size by party size
    palette="dark"
)
plt.title("Advanced Scatter Plot: Tips")
plt.grid(True)
plt.show()
```
