```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]

# Create a figure with 1 row and 2 columns of subplots
plt.figure(figsize=(9,5))

# First subplot - left side
plt.subplot(1, 2, 1)  # 1 row, 2 columns, subplot 1
plt.plot(x, y1, color='blue',linestyle='--')
plt.title("Plot 1")
plt.xlabel("common X value")
plt.ylabel("Plot of y1 axis")
plt.grid(True)
# Second subplot - right side
plt.subplot(1, 2, 2)  # 1 row, 2 columns, subplot 2
plt.plot(y1, x, color='black')
plt.title("Plot 2")
plt.xlabel("Common x value")
plt.ylabel("plot of y2 value")
plt.grid(True)
# Show all plots
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]
y3=[1,2,4,8,16]

# Create a figure with 1 row and 2 columns of subplots
plt.figure(figsize=(9,9))

# First subplot - left side
plt.subplot(2, 2, 1)  # 1 row, 2 columns, subplot 1
plt.plot(x, y1, color='blue',linestyle='--')
plt.title("Plot 1")
plt.xlabel("common X value")
plt.ylabel("Plot of y1 axis")
plt.grid(True)

# Second subplot - right side
plt.subplot(2, 2, 2)  # 1 row, 2 columns, subplot 2
plt.plot(x,y2 ,color='black')
plt.title("Plot 2")
plt.xlabel("Common x value")
plt.ylabel("plot of y2 value")
plt.grid(True)

# now plotting third row 
plt.subplot(2,2,3)
plt.plot(x,y3,linestyle='-.',color="red")
plt.xlabel("Common x value")
plt.ylabel("plot of y3 value")
plt.grid(True)
# Show all plots
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt

# Sample data — this represents scores of students out of 100
scores = [33, 45, 55, 55, 60, 61, 62, 67, 67, 69,
          70, 72, 74, 75, 75, 77, 78, 79, 80, 82,
          84, 85, 86, 88, 90, 92, 93, 94, 95, 100]

# Define bin edges manually to control ranges clearly
bins = [30, 50, 60, 70, 80, 90, 100]  # 6 bins

# Create histogram
plt.hist(scores, bins=bins, color='skyblue', edgecolor='black')

# Add titles and labels
plt.title("Student Score Distribution")
plt.xlabel("Score Range")
plt.ylabel("Number of Students")
plt.grid(True, linestyle='--', alpha=0.5)

# Display the histogram
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ["A", "B", "C", "D"]
sizes = [30, 20, 40, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = [0.2, 0, 0, 0]  # Explodes slice A

plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    explode=explode,
    autopct='%1.1f%%',       # Show percentages
    shadow=True,             # Add a shadow effect
    startangle=140           # Rotate the start of the pie
)

plt.title("Pie Chart Example")
plt.axis('equal')  # Equal aspect ratio ensures the pie is a circle.
plt.show()
```
