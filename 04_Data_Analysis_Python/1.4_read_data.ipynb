✅ How to Fix It: Add a Fake Browser Header

You can manually fetch the page HTML using requests with a fake user-agent header, then pass the HTML to pd.read_html.

```python
import pandas as pd
import requests
url ="https://en.wikipedia.org/wiki/Table_(information)"
headers={
    "user-agent":"Mozilla/5.0(Windows NT 10.0;Win64;x64 )"
}
response =requests.get(url,headers=headers)
html=response.text
# Now read HTML from the response
tables=pd.read_html(html)
# For example,print the first table
print(tables[0].head())
```
