```python
## Decorator
def main_welcome(func):
    def sub_welcome_method():
        print("Welcome to the advance python course ")
        func()
        print("Please learn these concepts properly")
    return sub_welcome_method
```

```python
## Decorators wth arguments 
def repeat(n):
    def decorator(func):
        def wrapper(*args,**kwargs):
            for _ in range(n):
                func(*args,**kwargs)
        return wrapper
    return decorator
```

```python
@repeat(3)
def say_hello():
    print("hello")
    
```

```python
say_hello()
```
